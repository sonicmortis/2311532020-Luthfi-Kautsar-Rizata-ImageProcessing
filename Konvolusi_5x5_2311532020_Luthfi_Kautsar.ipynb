```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("images.jpg").convert("L")
img_array = np.array(img)
```

```python
plt.imshow(img_array, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
kernel = np.ones((5,5)) / 25
print(kernel)

padded_img = np.pad(img_array, pad_width=2, mode='constant', constant_values=0)
```

```python
output = np.zeros_like(img_array)

for i in range(img_array.shape[0]):
    for j in range(img_array.shape[1]):
        region = padded_img[i:i+5, j:j+5]
        output[i,j] = np.sum(region * kernel)
```

```python
plt.imshow(output, cmap='gray')
plt.title("Filtered Image (5x5 Mean Filter)")
plt.axis('off')
```

```python
result_img = Image.fromarray(output)
result_img.save("hasil_konvolusi.jpg")
```
