```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_images(images, titles):
    plt.figure(figsize=(15, 5))
    for i in range(len(images)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.show()
```

```python
img = cv2.imread('Ranpo Edogawa.jpg', cv2.IMREAD_GRAYSCALE)

c_mul = 3
img_multiplied = np.clip(img.astype(np.float32) * c_mul, 0, 255).astype(np.uint8)
show_images([img, img_multiplied], ['Citra Asli', 'Hasil Perkalian (Kalibrasi)'])
```

```python
c_div = 100.0
img_divided = np.clip(img.astype(np.float32) / c_div, 0, 255).astype(np.uint8)
show_images([img, img_divided], ['Citra Asli', 'Hasil Pembagian (Normalisasi)'])
```
